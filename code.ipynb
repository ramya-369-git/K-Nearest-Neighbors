```python
import pandas as pd
```

```python
dataset=pd.read_csv(r"C:\Users\pramy\OneDrive\Desktop\supervised(1)\diabetes.csv")
df=pd.DataFrame(dataset)
df.head()
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df.drop(['SkinThickness','Age'],axis=1,inplace=True)
df.head(3)
```

```python
x=df.drop('Outcome',axis=1)
x=pd.DataFrame(x)
x.head(3)
```

```python
y=df['Outcome']
y=pd.DataFrame(y)
y.head(3)
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x_scale=ss.fit_transform(x)
x_scale_df=pd.DataFrame(x_scale,columns=x.columns)
x_scale_df.describe()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scale_df,y,test_size=0.2,random_state=42)
```

```python
import math
k=int(round(math.sqrt(len(y_test))))
if k%2==0:
    k+=1
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
model=KNeighborsClassifier(n_neighbors=k)
model.fit(x_train,y_train)
```

```python
y_test_pred=model.predict(x_test)
y_train_pred=model.predict(x_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
print("\nClassification Report:\n", classification_report(y_test, y_test_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_test_pred))
```

```python
print("Test Accuracy:", accuracy_score(y_test, y_test_pred))
print("Train Accuracy:", accuracy_score(y_train, y_train_pred))
```
